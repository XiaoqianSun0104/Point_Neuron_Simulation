# Case 1

```python
# necessary parameters

T = 100
dt = 0.1
gE_bar = 0.15
gI_bar = 0.30

Ne = 4
Ni = 4
N = Ne+Ni

maxns=1000


# customize any exc/inh attributes and create a list of neuron objects
exc_kwargs = {'externalInput':220}
inh_kwargs = {'tref':0.5, 'tau_m':15}
neuornObj_list = [neuron.Neuron(T, dt, gE_bar, gI_bar, 0, **exc_kwargs) for i in range(Ne)]+\
                [neuron.Neuron(T, dt, gE_bar, gI_bar, 1, **inh_kwargs) for i in range(Ni)]
print(len(neuornObj_list), 'neuron objects in total\n'); print()


# connectivity attributes
cm_kwargs = {'p_exc2exc':0.6, 'p_inh2exc':0.8, 'w_inh2exc':0.35}


# initialize simulator
simulator = simulation.Simulator(T, dt, N, Ne, Ni, maxns, neuornObj_list, **cm_kwargs)


print('Initial connectivity df with weights:'); print(simulator.CMW_df); print()
simuResult = simulator.run(ifVerbose=True)
print('The network generate spk:', simulator.get('netSpk'))
print('Final connectivity df with weights:'); print(simulator.CMW_df); print()
```
